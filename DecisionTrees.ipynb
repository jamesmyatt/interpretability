# Iris Dataset

```python
iris = load_iris()
X_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
y_iris = iris.target
```

```python
iris['feature_names']
```

```python
iris_dt = DecisionTreeClassifier()
iris_dt.fit(X_iris, y_iris)
```

```python
iris_dot_data = StringIO()
print('Features: ', iris.feature_names)
print('Targets: ', iris.target_names)
export_graphviz(iris_dt, out_file=iris_dot_data,
                class_names=iris.target_names,
                filled=True, rounded=True,
                feature_names=iris.feature_names,
                precision=1,
                special_characters=True)
iris_graph = pydotplus.graph_from_dot_data(iris_dot_data.getvalue())  
iris_graph.write_png('images/iris_decision_tree.png')
Image(iris_graph.create_png())
```
